```python
#2
side1=int(input('Enter side 1\t'))
side2=int(input('Enter side 2\t'))
side3=int(input('Enter side 3\t'))
print('sides of the Triangle:',side1,side2,side3)
if side1==side2==side3:
    print('equilateral triangle')
elif side1==side2 or side1==side3:
    print('isosceles triangle')
else:
    print('scalene triangle')
```

```python
#3
month=input('Enter month')
day=int(input('Enter date'))
if month=='march' or month=='March':
    if(day<20):
        print('Pisces (Fish)')
    else:
        print('Aries (Ram)')
elif month=='april'or month=='April':
    if(day<19):
        print('Aries (Ram)')
    else:
        print('Taurus (Bull)')
elif month=='may'or month=='May':
    if(day<20):
        print('Taurus (Bull)')
    else:
        print('Gemini (Twins)')
elif month=='june'or month=='June':
    if(day<21):
        print('Gemini (Twins)')
    else:
        print('Cancer (Crab)')
elif month=='july'or month=='July':
    if(day<22):
         print('Cancer (Crab)')
    else:
        print('Leo (Lion)')
elif month=='august'or month=='August':
    if(day<22):
        print('Leo (Lion)')
    else:
        print('Virgo (Virgin)')
elif month=='september'or month=='September':
    if(day<22):
        print('Virgo (Virgin)')
    else:
        print('Libra (Balance)')
elif month=='october'or month=='October':
    if(day<23):
         print('Libra (Balance)')
    else:
        print('Scorpius (Scorpion)')
elif month=='november'or month=='November':
    if(day<21):
        print('Scorpius (Scorpion)')  
    else:
        print('Sagittarius (Archer)')
elif month=='december'or month=='December':
    if(day<21):
        print('Sagittarius (Archer)')
    else:
        print('Capricornus (Goat)')
elif month=='january'or month=='January':
    if(day<19):
        print('Capricornus (Goat)')
    else:
        print('Aquarius (Water Bearer)')
elif month=='february'or month=='February':
    if(day<18):
        print('Aquarius (Water Bearer)')
    else:
        print('Pisces (Fish)')
```
